```python
import pandas as pd
import matplotlib

path='https://github.com/anastasiarazb/skillbox_nlp_demo/blob/master/%D0%9C%D1%81%D0%BA_5%D0%BB%D0%B5%D1%82.xls?raw=true'
data = pd.read_excel(path, skiprows=6)
```

```python
data
```

```python
data['date2'] = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)
data = data.sort_values('date2')


x = data['date2']
y = data['T']

plt.figure(figsize=(20, 5))
plt.plot(x, y, label='temperature')
plt.title('Температура в Москве')
plt.xlabel('index')
plt.ylabel('T')
plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day_first = pd.Timestamp(day=1, month=12, year=2016)
day_last = pd.Timestamp(day=1, month=1, year=2018)
condition = ((data['date2'] >= day_first) & (data['date2'] <= day_last))

data_small = data[condition]

x = data_small['date2']
y = data_small['T']

plt.figure(figsize=(20, 5))
plt.plot(x, y, label='temperature')
plt.title('Температура в Москве')
plt.xlabel('index')
plt.ylabel('T')
plt.legend()
```
